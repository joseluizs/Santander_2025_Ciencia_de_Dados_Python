```python
import pandas as pd
```

```python
# Primeiro passo é importar o dataframe
df = pd.read_csv('/content/dados_baguncados.csv')
display(df)
```

```python
# Segundo passo é padronizar as colunas str.strip() retirar espaços,
# str.lower() deixar todos os dados em minusculo, replace
df.columns = df.columns.str.strip().str.lower().str.replace(r'\s+', '_', regex=True)
print(df.columns)
```

```python
# Terceiro remoção de valores duplicados/duplicatas
df = df.drop_duplicates()
display(df)
```

```python
# Quarto remoção de colunas vazias
df = df.dropna(axis=1, how='all')
display(df)
```

```python
# Quinto tratar os tipos de dados no dataframe pd.to_datetime() para datas,
# pd.to_numeric para numeros, errors='coerce' gera valor/deixa vazio quando tem
# erro de data
df['data_venda'] = pd.to_datetime(df['data_venda'], errors='coerce')

df.info()
```

```python
# Sexto tratar linhas com valores vazios, como é numero vamos preencher com a media da coluna, passando o .fillna(df['quantidade'].mean())
df['quantidade'] = df['quantidade'].fillna(df['quantidade'].mean())
display(df)
```

```python
# Sétimo vamos padronizar valores de texto
df['uf'] = df['uf'].str.upper()
df['cliente_email'] = df['cliente_email'].str.lower()

display(df)
```

```python
# Criar colunas auxiliares com o dataframe
df['faturamento'] = df['quantidade'] * df['preco_unitario']
df
```

```python
df['cliente_mais_gastou'] = df['cliente_email'] + df['produto']
df
```

```python
# saber qual o cliente que mais gastou em compras e mostrar ele.
clientes_faturamento = df.groupby('cliente_email')['faturamento'].sum().reset_index()
cliente_mais_gastou = clientes_faturamento.loc[clientes_faturamento['faturamento'].idxmax()]
display(cliente_mais_gastou)
```
